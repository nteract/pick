```python
import json
from IPython.display import JSON, HTML
from unpack_har import unpack_messages
```

```python
with open("./kernel-messages.har") as h:
    messages = unpack_messages(json.load(h))
```

```python
messages[0]
```

```python
def simple_message_display(message):
    
    is_picky = "picky" in message["payload"]["metadata"] and message["payload"]["metadata"]["picky"]
    
    msg_type = message['payload']['header']['msg_type']
    
    content = ""
    # Looking to see if status messages are being sent by the subkernel and picky itself
    status_icon = "⛏" if is_picky else ""
    if msg_type == "status":
        content = message['payload']['content']['execution_state']
        if not is_picky:
            status_icon = "😈"
    
    icon = " "
    if message['type'] == 'receive':
        icon = f"""<span style="color: #3737e7">➡</span>"""
    else: 
        icon = f"""<span style="color: #37e737">←</span>"""
    
    
    return HTML(f"""<div style="font-family: monospace">
    <details>
    <summary>{icon} <b>{message['payload']['header']['msg_type']}</b> {status_icon}
    </summary>
    <pre>
    {json.dumps(message, indent=4)}
    </pre>
    </details>
    
    {content}
    </div>""")


for message in messages:
    display(simple_message_display(message))
```
